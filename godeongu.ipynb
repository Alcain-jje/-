```python
# 결측처리
for col in df.columns:
    df.loc[df[col] == 0, col] = None
```

```python
# 결측에 대한 선형 보간
for col in df.columns:
    if df[col].isna().sum() == 0:
        continue
    plt.figure(figsize=(10,5))
    plt.plot(df[col], alpha=0.4, linewidth=5)
    interploated = df[col].reset_index(drop=True).interpolate(method = 'cubic')
    interploated.index = df.index
    df[col] = interploated
    plt.plot(df[col], linewidth=0.5)
    plt.title(col)
    plt.show()
```

```python
def cross_corr_price_plotter(col):
    cross_corr = [df[['supply', col]].corr().iloc[0][1]] + [np.corrcoef(df.supply.iloc[:-i].values, df[col].iloc[i:].values)[0][1] for i in range(1, 4)]
    plt.figure(figsize=(3,4))
    plt.bar(range(0, 4), cross_corr, color='pink', width=0.5)
    plt.xlabel('Month')
    plt.ylabel('Corr')
    plt.ylim(-0.4, 0.6)
    plt.title(f"supply and {col}(+months)")
    plt.show()
```

```python
cross_corr_price_plotter('stock')
cross_corr_price_plotter('supply_price')
cross_corr_price_plotter('whole_price')
cross_corr_price_plotter('retail_price')
```
